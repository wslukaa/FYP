```python
from datetime import datetime

hts = {}
with open('Trend_hashtag_jsonlines') as f:
    for i, l in enumerate(f):
        if i < 1:
            continue
        row = json.loads(l)
        actual = int(row[94])
        #print(actual)
        ht = row[1]
        #ts = row[2]
        ts = datetime.strptime(row[2],"%Y%m%d-%H%M%S")
        if ht not in hts:
            hts[ht] = []
        hts[ht].append([ts,actual])
        

cluster = {}
with open('clusterfyp_results/clusterfyp_withpca_optimised_562cluster3words.txt') as f:
    for l in f:
        lt = json.loads(l)
        for c in lt:
            c0 = '#' + c[0].strip()
            c1 = '#' + c[1].strip()
            c2 = '#' + c[2].strip()
            cluster_id = c0 + c1 +c2
            for cc in [c0,c1,c2]:
                if cc in hts:
                    for data in hts[cc]:
                        if data[1] == 1:
                            if cluster_id not in cluster:
                                cluster[cluster_id] = {}
                            if cc not in cluster[cluster_id]:
                                cluster[cluster_id][cc] = []
                            cluster[cluster_id][cc].append(data[0].strftime("%Y%m%d-%H%M%S"))

print(len(cluster))
count = 0
for k,v in cluster.items():
    for kk, vv in v.items():
        count += len(vv)
print(count)
print(cluster)

#with open('timestamp_hashtag_istrend_1_according_to_pca562cluster3words','a') as f:
#    f.write(json.dumps(cluster))
    
            
```
