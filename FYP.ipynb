```python
import json
from datetime import datetime
import matplotlib.pyplot as plt

with open("stream_tweets-20171116-012802.txt", "r") as f:
    x = []
    y = []
    for line in f:
        d = json.loads(line)
        x.append(datetime.strptime(d['created_at'], '%a %b %d %H:%M:%S %z %Y'))
        y.append(d['user_friends_count'])
    plt.plot(x, y)
    plt.gcf().autofmt_xdate()
    plt.show()
    
```
